Step 1: import libraries and set dataframe

```python
import numpy as np 
import pandas as pd 
```

```python
df = pd.read_csv('../data/netflix_titles.csv')
df.head()
```

Step 2: get info from data

```python
df.info()
df.nunique()
df.isnull().sum()
df.sample(5)
```

Step 3: Cleaning

```python
# Handle missing values

df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')

# Convert date_added to proper datetime format
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
# Add year and month of release 
df['added_year'] = df['date_added'].dt.year
df['added_month'] = df['date_added'].dt.month

```

```python
# Extract duration as integer (Represents the runtime in minutes for movies, the amount of seasons for TV shows)

df['duration_int'] = (df['duration'].fillna('0')).str.extract('(\d+)').astype(int)
```

```python
df.head()
```

```python
#Drop duplicates
df = df.drop_duplicates()
```

```python
# Store genres and countries as list for easier searching
df['genre_list'] = df['listed_in'].str.split(',').apply(lambda x: [i.strip() for i in x] if isinstance(x, list) else [])
df['country_list'] = df['country'].str.split(', ')
df.head()
```

```python
# Save 
df.to_csv('../data/netflix_clean.csv', index=False)
```
