```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('../data/netflix_clean.csv')
df.head()
```

```python
df['genre_list'] = df['listed_in'].str.split(',').apply(lambda x: [i.strip() for i in x] if isinstance(x, list) else [])
df.head()
```

```python
df2 = df.copy()
df2 = df2.explode('genre_list')
df2.head()
df2.save
```

```python
# Which months have the highest amount of releases

moviesPerMonth = df.groupby(['added_month', 'type']).size().unstack(fill_value=0)
plt.figure(figsize=(12, 6))
moviesPerMonth.plot(kind='bar', stacked=True, figsize=(12,6))

plt.xlabel("Month")
plt.ylabel("Number Of Releases")
plt.title("Which months have the most new releases?")

plt.xticks(rotation=0)
plt.legend(title="Type")
plt.savefig('../visuals/NewReleases.png')
plt.show()
```

It seems that content releases are scheduled out fairly evenly, with the most amount of content being added in July and the least being added in February. 

Movie/TV distribution:

```python


type_counts = df['type'].value_counts()
percentages = ((type_counts / type_counts.sum() * 100).round(1))

plt.figure(figsize=(12, 12))
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', startangle=90)

plt.suptitle('Movies vs. TV distribution on Netflix', fontsize=14, fontweight='bold')
plt.axis('Equal')
plt.savefig('../visuals/MovieTVDistribution.png')
plt.show()
```

```python
# Split genres into individual rows
tv = df[df['type'] == 'TV Show'].copy()
tv_genres = tv.explode('genre_list')

genre_counts = tv_genres['genre_list'].value_counts().head(15)

plt.figure(figsize=(14,6))
plt.barh(genre_counts.index, genre_counts.values)
plt.gca().invert_yaxis()
plt.xlabel("Count")
plt.title("Most Common TV Show Genres")
plt.savefig('../visuals/CommonTVShows.png')
plt.show()

```

It seems that, aside from "International" TV shows, the most commonly released genres of TV show are Dramas, followed by comedies, crime, and kid's shows. 

```python
movies = df[df['type'] == 'Movie'].copy()
movie_genres = movies.explode('genre_list')
genre_counts_movies = movie_genres['genre_list'].value_counts().head(15)

plt.figure(figsize=(15,6))
plt.barh(genre_counts_movies.index, genre_counts_movies.values)
plt.gca().invert_yaxis()

plt.xlabel("Count")
plt.title("Most Common Movie Genres")
plt.savefig('../visuals/Common Movie Genres.png')
plt.show()
```

Dramas and Comedies dominate movie genres as well, followed by documentaries and action/adventure

```python
yearly_counts = df.groupby('added_year').size()

plt.figure(figsize=(12, 6))
plt.plot(yearly_counts.index, yearly_counts.values, marker='o')
plt.xlabel("Year")
plt.xlim(left=2014)
plt.ylabel("Number of Releases")
plt.title("Netflix Content Released per year")
plt.grid(True)
plt.savefig('../visuals/ContentReleasedPerYear.png')
plt.show()
```

It seems that Netflix content has been steadily increasing in quantitity since 2014, but then experienced a decline beginning in 2019 that then sharpened in 2020, likely due to the pandemic.
